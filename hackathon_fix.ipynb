```python
import re
import string

# Fungsi untuk menghapus URL
def remove_URL(text):
    if text is not None and isinstance(text, str):
        url = re.compile(r'https?://\S+|www\.\S+')
        return url.sub(r'', text)
    return text

# Fungsi untuk menghapus username (@username)
def remove_username(text):
    if text is not None and isinstance(text, str):
        return re.sub(r'@[A-Za-z0-9_]+', '', text)
    return text

# Fungsi untuk menghapus HTML
def remove_html(text):
    if text is not None and isinstance(text, str):
        html = re.compile(r'<.*?>')
        return html.sub(r'', text)
    return text

# Fungsi untuk menghapus emoji
def remove_emoji(text):
    if text is not None and isinstance(text, str):
        emoji_pattern = re.compile("["
            u"\U0001F600-\U0001F64F"
            u"\U0001F300-\U0001F5FF"
            u"\U0001F680-\U0001F6FF"
            u"\U0001F700-\U0001F77F"
            u"\U0001F780-\U0001F7FF"
            u"\U0001F800-\U0001F8FF"
            u"\U0001F900-\U0001F9FF"
            u"\U0001FA00-\U0001FA6F"
            u"\U0001FA70-\U0001FAFF"
            u"\U0001F004-\U0001F0CF"
            u"\U0001F1E0-\U0001F1FF"
        "]+", flags=re.UNICODE)
        return emoji_pattern.sub(r'', text)
    return text

# Fungsi untuk menghapus simbol
def remove_symbols(text):
    if text is not None and isinstance(text, str):
        return re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text

# Fungsi untuk menghapus angka
def remove_numbers(text):
    if text is not None and isinstance(text, str):
        return re.sub(r'\d', '', text)
    return text

# Proses cleaning bertahap
df['cleaning'] = df['Review Text'].apply(remove_URL)
df['cleaning'] = df['cleaning'].apply(remove_username)
df['cleaning'] = df['cleaning'].apply(remove_html)
df['cleaning'] = df['cleaning'].apply(remove_emoji)
df['cleaning'] = df['cleaning'].apply(remove_symbols)
df['cleaning'] = df['cleaning'].apply(remove_numbers)

df[['Review Text', 'cleaning']].head()
```

```python
import pandas as pd

# Membaca kamus kata baku dari file CSV di folder Data
kamus_df = pd.read_excel('C:/Users/madea/Desktop/Kelas_Bagus_3/scraping/kamus_kata_baku.xlsx')
kamus_tidak_baku = dict(zip(kamus_df['tidak_baku'], kamus_df['kata_baku']))

def replace_taboo_words(text, kamus_tidak_baku):
    if isinstance(text, str):
        words = text.split()
        replaced_words = []
        kalimat_baku = []
        kata_diganti = []
        kata_tidak_baku_hash = []

        for word in words:
            if word in kamus_tidak_baku:
                baku_word = kamus_tidak_baku[word]
                if isinstance(baku_word, str) and all(char.isalpha() for char in baku_word):
                    replaced_words.append(baku_word)
                    kalimat_baku.append(baku_word)
                    kata_diganti.append(word)
                    kata_tidak_baku_hash.append(hash(word))
                # else:
                #     replaced_words.append('<karakter eksplisit>')
                #     kalimat_baku.append(word)
            else:
                replaced_words.append(word)
        replaced_text = ' '.join(replaced_words)
    else:
        replaced_text = ''
        kalimat_baku = []
        kata_diganti = []
        kata_tidak_baku_hash = []

    return replaced_text, kalimat_baku, kata_diganti, kata_tidak_baku_hash

# Terapkan normalisasi pada kolom 'case_folding'
df[['normalized_text', 'kalimat_baku', 'kata_diganti', 'kata_tidak_baku_hash']] = df['case_folding'].apply(lambda x: pd.Series(replace_taboo_words(x, kamus_tidak_baku)))

df[['case_folding', 'normalized_text']].head()
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from functools import lru_cache
from tqdm import tqdm
import pandas as pd

# Inisialisasi stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Aktifkan tqdm untuk apply
tqdm.pandas()

# Caching stemmer per kata
@lru_cache(maxsize=100000)
def cached_stem(word):
    return stemmer.stem(word)

# Fungsi stemming per ulasan
def stem_text(tokens):
    if not isinstance(tokens, list):
        return ''
    return ' '.join([cached_stem(word) for word in tokens if isinstance(word, str) and word.strip() != ''])

# Terapkan ke DataFrame dengan progress bar
df['stemming_data'] = df['stopword_removal'].progress_apply(stem_text)

# Tampilkan hasil awal
df.head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
from collections import Counter

# Daftar stopwords tambahan (disesuaikan dengan cell sebelumnya)
custom_stopwords = [
    'https', 'co', 'RT', '...', 'amp', 'lu', 'deh', 'fyp', 'ya', 'gue', 'cs', 'kayak', 'nih',
    'sih', 'yg', 'nya', 'aja', 'sdh', 'gak', 'ga', 'update', 'apk', 'ajar', 'si',
    'dong', 'banget', 'mantap', 'wkwk', 'wkwkwk', 'anjay', 'anjir', 'fix', 'auto', 'gameplay',
    'grafik', 'grafis', 'fitur', 'akun', 'login', 'server', 'rank', 'match', 'hero', 'skin',
    'diamond', 'mlbb', 'ml', 'lag', 'bug', 'coba', 'main', 'mainin', 'udah', 'dah', 'baru',
    'lama', 'bikin', 'bisa', 'kayak', 'gitu', 'gini', 'jadi', 'dulu', 'malah', 'terus', 'makin',
    'parah', 'tolong', 'please', 'dev', 'developer', 'moonton', 'playstore', 'download', 'install',
    'uninstall', 'crash', 'force', 'close', 'error', 'loading', 'tunggu', 'cepat', 'slow', 'buruk',
    'baik', 'bagus', 'jelek', 'nice', 'cool', 'great', 'bad', 'good', 'very', 'really', 'just',
    'like', 'love', 'hate', 'fun', 'boring', 'seru', 'nggak', 'enggak', 'tdk', 'tidak', 'iya',
    'oke', 'ok', 'no', 'yes'
 ]

def plot_ngram_frequencies(texts, ngram_range=(4,4), top_n=10, color='lightcoral'):
    vectorizer = CountVectorizer(ngram_range=ngram_range, stop_words=custom_stopwords)
    X = vectorizer.fit_transform(texts)

    # Hitung total frekuensi setiap n-gram
    ngram_counts = X.sum(axis=0).A1
    ngram_vocab = vectorizer.get_feature_names_out()
    freq_dist = dict(zip(ngram_vocab, ngram_counts))

    # Ambil n-gram dengan frekuensi tertinggi
    top_ngrams = Counter(freq_dist).most_common(top_n)
    labels, values = zip(*top_ngrams)

    # Visualisasi horizontal bar chart dengan style yang konsisten
    plt.figure(figsize=(12, 7))
    bars = plt.barh(labels[::-1], values[::-1], color=color, edgecolor='black')
    plt.title(f'Top {top_n} {ngram_range[0]}-gram Frekuensi pada Review Mobile Legends', fontsize=16, fontweight='bold')
    plt.xlabel('Frekuensi', fontsize=13)
    plt.ylabel(f'{ngram_range[0]}-gram', fontsize=13)

    # Tambahkan label angka di ujung bar
    for bar in bars:
        width = bar.get_width()
        plt.text(width + max(values)*0.01, bar.get_y() + bar.get_height()/2, f'{int(width)}', va='center', fontsize=12)

    plt.grid(axis='x', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

# Visualisasi 4-gram pada kolom 'stemming_data'
plot_ngram_frequencies(df['stemming_data'], ngram_range=(4,4), top_n=10, color='lightcoral')
```

```python
import re
import pandas as pd

# 1) Ambil leksikon (positif/negatif) dan pastikan lowercase
positive_url = "https://raw.githubusercontent.com/fajri91/InSet/master/positive.tsv"
negative_url = "https://raw.githubusercontent.com/fajri91/InSet/master/negative.tsv"

positive_lexicon = set(pd.read_csv(positive_url, sep="\t", header=None)[0].str.lower())
negative_lexicon = set(pd.read_csv(negative_url, sep="\t", header=None)[0].str.lower())

# 2) Tentukan kandidat kolom teks (urutkan dari yang paling kamu harapkan)
candidate_cols = [
    "steming_data", "stemming_data", "stemmed_text", "stem_text",
    "normalized_text", "clean_text", "case_folding", "text", "tweet", "content"
]

# Pilih kolom pertama yang ada di data
text_col = next((c for c in candidate_cols if c in data.columns), None)
if text_col is None:
    raise ValueError(
        f"Tidak ditemukan kolom teks. Kolom yang tersedia: {list(data.columns)}.\n"
        "Ganti 'candidate_cols' atau beri tahu saya nama kolom yang benar."
    )

# 3) Fungsi hitung sentimen (tokenisasi simpel + lowercase)
def determine_sentiment(text: str):
    if isinstance(text, str):
        # lowercase + ambil token alfanumerik
        tokens = re.findall(r"\w+", text.lower(), flags=re.UNICODE)

        pos = sum(1 for w in tokens if w in positive_lexicon)
        neg = sum(1 for w in tokens if w in negative_lexicon)

        score = pos - neg
        sentiment = "Positif" if score > 0 else "Negatif"
        return score, sentiment
    return 0, "Negatif"

# 4) Terapkan ke kolom terpilih
data[["Score", "Sentiment"]] = data[text_col].apply(
    lambda x: pd.Series(determine_sentiment(x))
)

# Lihat hasil
print(text_col, "dipakai sebagai kolom teks.")
data[[text_col, "Score", "Sentiment"]].head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# 1) Tentukan kolom teks yang ada
candidate_cols = [
    "steming_data", "stemming_data", "stemmed_text", "stem_text",
    "normalized_text", "clean_text", "case_folding", "text", "tweet", "content"
]
text_col = next((c for c in candidate_cols if c in data.columns), None)
if text_col is None:
    raise ValueError(f"Tidak ketemu kolom teks. Kolom tersedia: {list(data.columns)}")

# 2) Pastikan kolom 'Sentiment' ada
if "Sentiment" not in data.columns:
    raise ValueError("Kolom 'Sentiment' belum ada. Jalankan langkah penentuan sentimen dulu.")

# 3) Bersihkan data dan ambil X, y
cleaned_data = data.dropna(subset=[text_col]).copy()
X = cleaned_data[text_col].astype(str)
y = cleaned_data["Sentiment"].astype(str)

print(f"Kolom teks yang dipakai: {text_col}")

# 4) Split data (fallback tanpa stratify jika komposisi kelas tidak memungkinkan)
try:
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.1, random_state=42, stratify=y
    )
except ValueError:
    # Misal salah satu kelas terlalu sedikit
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.1, random_state=42, stratify=None
    )
    print("Peringatan: stratify dimatikan karena distribusi kelas tidak memadai.")

print("Jumlah data latih:", len(X_train))
print("Jumlah data uji:", len(X_test))
print('===========================')

# 5) Vectorize
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec  = vectorizer.transform(X_test)

X_train_vec.shape, X_test_vec.shape
```

```python
import matplotlib.pyplot as plt


train_size = len(X_train)
test_size = len(X_test)

plt.figure(figsize=(6, 4))
bars = plt.bar(['Data Training', 'Data Testing'], [train_size, test_size], color=['blue', 'orange'])

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f'{height} ({height / (train_size + test_size) * 100:.2f}%)',
             ha='center', va='bottom')

plt.title('Jumlah Data Latih dan Data Uji')
plt.xlabel('Jenis Data')
plt.ylabel('Jumlah Data')
plt.show()
```

# ***SVM (Support Vector Machine), KNN (K-Nearest Neighbor), Naive Bayes, Random Forest, Decision Tree, Neural Network***

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.neural_network import MLPClassifier
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding, SpatialDropout1D
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier

# Initialize models
models = {
    "SVM": SVC(kernel='linear', random_state=42),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Naive Bayes": MultinomialNB(),
    "Random Forest": RandomForestClassifier(random_state=42, n_estimators=100),
    "Decision Trees": DecisionTreeClassifier(random_state=42),
    "Neural Network": MLPClassifier(random_state=42, hidden_layer_sizes=(100,), max_iter=500),
}

# Train models
results = {}
for model_name, model in models.items():
    model.fit(X_train_vec, y_train)
    y_pred = model.predict(X_test_vec)
    results[model_name] = {
        "accuracy": accuracy_score(y_test, y_pred),
        "classification_report": classification_report(y_test, y_pred, output_dict=True),
        "confusion_matrix": confusion_matrix(y_test, y_pred)
    }
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for model_name, result in results.items():
    fig, ax = plt.subplots(figsize=(4, 4))
    sns.heatmap(
        result["confusion_matrix"],
        annot=True,
        fmt='d',
        cmap="YlGnBu",
        cbar=False,
        xticklabels=['Negatif','Positif'],
        yticklabels=['Negatif','Positif'],
        ax=ax,
        square=True,
        linewidths=0.5
    )
    print(f"\nConfusion Matrix for {model_name}:")
    ax.set_title(f"{model_name} Confusion Matrix", fontsize=12)
    ax.set_xlabel("Predicted", fontsize=12)
    ax.set_ylabel("Actual", fontsize=12)


    plt.tight_layout()
    plt.show()
```
